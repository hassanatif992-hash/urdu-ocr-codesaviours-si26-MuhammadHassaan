## What is OCR?

OCR (Optical Character Recognition) is a technology used to convert text from images into machine-readable text. It helps computers recognize printed or handwritten text from scanned documents, books, and images. OCR is widely used for digitizing paper documents.

## Why is Urdu OCR harder than English OCR?

Urdu OCR is harder because Urdu is written in a cursive style where letters connect together. The shape of letters changes depending on their position in a word. Urdu also contains dots and complex ligatures, making text recognition more difficult.

## Real-world uses of Urdu OCR

Urdu OCR is useful for digitizing old Urdu books, newspapers, and historical records. It can help offices convert scanned Urdu documents into editable text. It is also useful for extracting text from signboards and printed forms.
